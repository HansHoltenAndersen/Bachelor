```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
I = 350.0      # income
px = 2.0       # original price of housing
py = 1.0       # price of other goods
a = 0.6        # Cobb-Douglas share for housing

# New budget with higher px (pivot inward on x-axis)
px2 = 3.5

# Optimal bundles (Cobb-Douglas: x = a*I/px, y = (1-a)*I/py)
x1 = a * I / px
y1 = (1 - a) * I / py

x2 = a * I / px2
y2 = (1 - a) * I / py

# Utility levels
U1 = (x1**a) * (y1**(1 - a))
U2 = (x2**a) * (y2**(1 - a))

# Budget lines
x_line1 = np.linspace(0, I / px, 200)
y_line1 = (I - px * x_line1) / py

x_line2 = np.linspace(0, I / px2, 200)
y_line2 = (I - px2 * x_line2) / py

# Grid for indifference curves
x = np.linspace(0.001, I / min(px, px2) * 1.15, 400)
y = np.linspace(0.001, I / py * 1.05, 400)
X, Y = np.meshgrid(x, y)
U = (X**a) * (Y**(1 - a))

# Plot
plt.figure(figsize=(7.5, 6))

# Budget lines
plt.plot(x_line1, y_line1, label="Budget line (px=2)")
plt.plot(x_line2, y_line2, linestyle="--", label="New budget (px=3.5)")

# Indifference curves through optima
plt.contour(X, Y, U, levels=[U1], linewidths=1.8, colors='C1')
plt.contour(X, Y, U, levels=[U2], linewidths=1.8, colors='C2', linestyles="dotted")

# Optima points
plt.scatter([x1], [y1], s=55, label="Optimum A", color='C1')
plt.scatter([x2], [y2], s=55, marker="s", label="Optimum B", color='C2')

# Vertical dotted lines down to x-axis (Housing consumption)
plt.vlines([x1, x2], ymin=0, ymax=[y1, y2], linestyles="dotted", colors="gray")

# Annotations for points A and B
plt.annotate("A", (x1, y1), xytext=(5, 5), textcoords="offset points")
plt.annotate("B", (x2, y2), xytext=(5, -12), textcoords="offset points")

# Labels and limits
plt.xlabel("Housing")
plt.ylabel("Other consumption")
plt.title("Price increase ")
plt.xlim(0, max(x_line1.max(), x_line2.max()) * 1.1)
plt.ylim(0, (I / py) * 1.05)
plt.legend(loc="upper right")
plt.tight_layout()

plt.show()
```
